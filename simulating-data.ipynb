I spent a long time trying to find the best way to get my variable data to relate to eachother, I wanted to use numpy.choice to add probability to each age group, I decided the best way to do this was by row. [thispointer.com](https://thispointer.com/pandas-count-rows-in-a-dataframe-all-or-those-only-that-satisfy-a-condition/) helped me a great deal along with pandas documentation. 
[pandas.DataFrame.apply](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html)
Applies a function along an axis of the DataFrame.




```python
#creating a label function to choose the probability alongside numpy.choice 
UHC = 'yes','no'
def label_UHC (row):
   if row['Ages'] == '0-14' :
      return np.random.choice(UHC, 1, p = [0.07, 0.93])
   if row['Ages'] == '15-24' :
      return np.random.choice(UHC, 1, p = [0.25, 0.75])
   if row['Ages'] == '25-44' :
      return np.random.choice(UHC, 1, p = [0.35, 0.65])
   if row['Ages'] == '45-64' :
      return np.random.choice(UHC, 1, p = [0.5, 0.5])
   if row['Ages'] == '65-74' :
      return np.random.choice(UHC, 1, p = [0.70, 0.30])
   if row['Ages'] == '75+' :
      return  np.random.choice(UHC, 1, p = [0.87, 0.13])
    

 
```

```python
dfsample['Underlying Health Condition'] = dfsample.apply (lambda row: label_UHC(row), axis=1)
dfsample

```

I had noticed bracets around my yes and no's and could not figure out why, after much research and studying the [pandas documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html) and changing the code around many times I Found this great source on [stackoverflow.com](https://stackoverflow.com/questions/38147447/how-to-remove-square-bracket-from-pandas-dataframe) 

```python
dfsample['Underlying Health Condition'] = dfsample['Underlying Health Condition'].str[0]
dfsample
```

# Recovered


Most people who get Covid-19 wil recover without complications, the chances of recovery decreases with age and underlying health conditions. For this variable I did a simple line of code, if the person with the case died, they did not recover. 


```python
def label_rec (row):
   if row['Died'] == 'yes' :
      return 'no'
   if row['Died'] == 'no' :
      return 'yes'
    


```

```python
dfsample['Recovered'] = dfsample.apply (lambda row: label_rec(row), axis=1)
dfsample
```
